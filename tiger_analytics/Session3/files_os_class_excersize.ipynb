```python
import csv
import os
```

[dataset link](https://drive.google.com/file/d/1WShGELaUTe_NsMvLfDl10OVQJgNgkFwg/view?usp=sharing)

```python
path = "/Users/gunnvantsaini/Downloads/pollution"
file_names = os.listdir(path)
```

```python
sensor_map = {}
for file_name in file_names:
    path_csv = os.path.join(path,file_name)
    with open(path_csv,"r") as infile:
        reader = csv.DictReader(infile,delimiter=",")
        row1 = next(reader)
        long = row1['longitude']
        lat = row1['latitude']
        lat_long = str(lat)+"_"+str(long)
        if lat_long not in sensor_map:
            sensor_map[lat_long] = []
            sensor_map[lat_long].append(file_name)
        else:
            sensor_map[lat_long].append(file_name)
```
